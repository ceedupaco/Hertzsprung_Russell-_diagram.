## Diagrama Hertzsprung–Russell para luminosidad en función de la temperatura

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data/hygdata_v3-1.csv')[['ci','lum']]#se importan los datos, tomando las columnas necesarias
```

Para obtener los valores de la temperatura se usa la expresion $ B-V= \frac{7200}{T-0.64}$ derivada de la aproximación de Wien a la ley de Planck.

```python
df['tem'] = (-7200/(df.ci +0.4)) 
```

Para mejorar la visualización tomamos el logaritmo de la luminosidad

```python
df['logl']= np.log(df.lum)
```

```python
df.to_csv('dftemperatura.csv')#se guarda el nuevo dataset
```

```python
df2 = pd.read_csv('dftemperatura.csv')[['logl', 'tem','ci']]#se importa nuevamente
```

Primera aproximación al esquema a blanco y negro

```python
fig = plt.figure(figsize=(6, 8),dpi=100)
ax = fig.add_axes([.1, .1, .85, .8])
ax.set_title('Hertzsprung-Russell Diagram', fontsize=18)
ax.title.set_position([.5, 1.03])
ax.set_xlabel('Tempetatura K')
ax.set_ylabel('luminosity')

ax.scatter(-df2['tem'],df2['logl'],marker='.',s=[5] * len(df2),facecolors='black',linewidth=0)
ax.set_xlim(40000, 0)
```

Se usa la misma máscara de índice de color del diagrama anterior

```python
def bv2rgb(bv):
    t = (5000 / (bv + 1.84783)) + (5000 / (bv + .673913))
    x, y = 0, 0
    
    if 1667 <= t <= 4000:
        x = .17991 - (2.66124e8 / t**3) - (234358 / t**2) + (877.696 / t)
    elif 4000 < t:
        x = .24039 - (3.02585e9 / t**3) + (2.10704e6 / t**2) + (222.635 / t)
        
    if 1667 <= t <= 2222:
        y = (-1.1063814 * x**3) - (1.34811020 * x**2) + 2.18555832 * x - .20219683
    elif 2222 < t <= 4000:
        y = (-.9549476 * x**3) - (1.37418593 * x**2) + 2.09137015 * x - .16748867
    elif 4000 < t:
        y = (3.0817580 * x**3) - (5.87338670 * x**2) + 3.75112997 * x - .37001483
        
    X = 0 if y == 0 else x / y
    Z = 0 if y == 0 else (1 - x - y) / y
    
    r, g, b = np.dot([X, 1., Z],
        [[3.2406, -.9689, .0557], [-1.5372, 1.8758, -.204], [-.4986, .0415, 1.057]])
    
    R = np.clip(12.92 * r if (r <= 0.0031308) else 1.4 * (r**2 - .285714), 0, 1)
    G = np.clip(12.92 * g if (g <= 0.0031308) else 1.4 * (g**2 - .285714), 0, 1)
    B = np.clip(12.92 * b if (b <= 0.0031308) else 1.4 * (b**2 - .285714), 0, 1)
    
    return [R, G, B, np.random.ranf()]

color = df2['ci'].apply(bv2rgb)
```

```python
fig = plt.figure(
    figsize=(6, 8),
    facecolor='black',
    dpi=100)
ax = fig.add_axes([.1, .1, .85, .8])
ax.set_axis_bgcolor('black')
ax.set_title('Hertzsprung-Russell Diagram ', color='white', fontsize=18)
plt.text(0,-12 , 'CEPC',color='white', fontsize = 10, horizontalalignment='center', verticalalignment='center') 
ax.title.set_position([0.3, 1.03])
ax.set_xlabel('Temperature [K]', color='white')
ax.set_ylabel('Log(Luminosity)', color='white')
ax.scatter(-df2['tem'],df2['logl'],marker='.',s=[5] * len(df2),facecolors=color,linewidth=0)
ax.set_xlim(40000, -1000)
ax.tick_params(top='off', right='off', direction='out', colors='white')


#Áreas correspondientes a los tipos de estrella
x1 = [22000]
y1 = [-2]
x2 = [30000]
y2 = [15]
x3 = [3000,]
y3 = [15,]
x4 = [2500]
y4 = [5]
colors1 = np.random.rand(4)
colors2 = np.random.rand(4)
colors3 = np.random.rand(4)
colors4 = np.random.rand(4)

plt.scatter(x1, y1, s=5000, c=colors1, alpha=0.4, marker='o',linewidth=2)
plt.scatter(x2, y2, s=20000, c=colors2, alpha=0.4, marker='o',linewidth=2)
plt.scatter(x3, y3, s=20000, c=colors3, alpha=0.4, marker='o',linewidth=2)
plt.scatter(x4, y4, s=7000, c=colors4, alpha=0.4, marker='o',linewidth=2)


plt.text(29000, 15 , 'Blue Giants',color='white', fontsize = 10, horizontalalignment='center', verticalalignment='center') 
plt.text(22000, -2 , 'White Dwarfs ',color='white', fontsize = 10, horizontalalignment='center', verticalalignment='center') 
plt.text(3000, 15 , 'Red supergiants',color='white', fontsize = 10, horizontalalignment='center', verticalalignment='center') 
plt.text(2500, 5 , 'Red Giants',color='white', fontsize = 10, horizontalalignment='center', verticalalignment='center') 
plt.text(5000, -5 , 'Main Sequence',color='white', fontsize = 10, horizontalalignment='center', verticalalignment='center') 
plt.text(30000, 5 , 'Main Sequence',color='white', fontsize = 10, horizontalalignment='center', verticalalignment='center') 


```

Temperaturas

Las estrellas son enormes aglomeraciones de gas, principalmente Hidrógeno, cuya temperatura es tan alta, debido
a la fusión de este elemento, que irradian luz a lo largo de todo el espectro electromagnético. 
Poseen diferentes temperaturas que varían desde los 2000 grados Celsius hasta los 50000.

De la misma forma que al calentar una pieza de metal cambia de color, al principio rojo, 
luego amarillo hasta llegar al blanco, el color de una estrella varía según su temperatura superficial. 
Las estrellas más frías son las rojas, y las más calientes son las azules. Estos colores suelen percibirse a 
simple vista, por ejemplo, Antares (la estrella principal de Scorpius) que es de color rojo, o Rigel (en Orion) 
de color azul. En astronomía se utiliza la escala Kelvin para indicar temperaturas, donde el cero absoluto 
es -273 grados Celsius.
Para definir el color de una estrella, Johnson y Morgan (1950), crearon el sistema UBV (del inglés Ultravioleta, Azul, Visible). Las mediciones se realizaban mediante un fotómetro fotoeléctrico para medir la intensidad de la radiación en longitudes de onda específicas.
